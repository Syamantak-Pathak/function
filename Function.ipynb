**5. What are iterators in Python and how do they differ from iterables?**
- *Iterable*: An object that can return an iterator using iter(). Examples: list, tuple, string.
- *Iterator*: An object with a _next_() method to fetch next value and StopIteration ends it.
  - Example (Iterable):
numbers = [1, 2, 3]  # list is iterable
for n in numbers:
    print(n)
  - Example (Iterator):
numbers = iter([1, 2, 3])  # converting to iterator
print(next(numbers))  # 1
print(next(numbers))  # 2

**7. What are the advantages of using generators over regular functions?**
- *Generators :*
  - Use less memory â€“ values are generated one at a time
  - Faster with large datasets
  - Maintain state between calls
- *Regular Functions:*
  - Store all values in memory at once
  - Can be slower with large data  
  - Start from scratch every time
- *Example:*
  - Generator:
def gen_numbers():
    for i in range(1000000):
     yield i
  - List (Regular function alternative):
def list_numbers():
    return [i for i in range(1000000)]
- Generator is more efficient when iterating over a large range.

**10.What is the difference between map(), reduce(), and filter() functions in Python?**
- *map() Function*
  - Purpose:** Transforms (or maps) each element of an iterable using a specified function.
  - Input: A function and an iterable (like a list, tuple, etc.)
  - Output: A new iterable (map object) where each element is the result of applying the function to the corresponding element of the original iterable.
  - Example: map(lambda x: x**2, [1, 2, 3]) # Output: [1, 4, 9]
- *filter() Function*
  - Purpose: Filters out elements from an iterable based on a condition defined in a function.
  - Input: A function that returns True or False, and an iterable.
  - Output: A new iterable (filter object) containing only the elements for which the function returned True.
  - Example: filter(lambda x: x % 2 == 0, [1, 2, 3, 4]) # Output: [2, 4]
- *reduce() Function*
  - Purpose: Reduces a sequence to a single cumulative value by applying a binary function cumulatively to the items.
  - Input: A function that takes two arguments, and an iterable.
  - Output: A single result, not a list or iterable.
  - Example Use Case: from functools import reduce
reduce(lambda x, y: x + y, [1, 2, 3, 4]) #Output: 10

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def expo(n):
  n=int(input("Enter the range:"))
  for i in range(n+1):
    yield 2**i
e=expo(5)
for v in e:
  print(v)
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms  # total number of terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            next_term = self.a + self.b
            self.a, self.b = self.b, next_term
            self.count += 1
            return next_term


fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```
